```python
#Пример получения осей для subplots:

# Two subplots, unpack the axes array immediately
X = np.linspace(-2,2,100)

Y = X**3


f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(X, Y)

ax1.set_title('Sharing Y axis')

ax1.spines['left'].set_position('center')

ax1.spines['bottom'].set_position('center')

ax1.spines['top'].set_visible(False)

ax1.spines['right'].set_visible(False)



ax2.scatter(X, Y)

ax2.spines['left'].set_position('center')

ax2.spines['bottom'].set_position('center')

ax2.spines['top'].set_visible(False)

ax2.spines['right'].set_visible(False)



f, axarr = plt.subplots(2, 2) # Four axes,
                               # returned as a 2-d array



axarr[0, 0].plot(X, Y)

axarr[0, 0].spines['left'].set_position('center')

axarr[0, 0].spines['bottom'].set_position('center')

axarr[0, 0].spines['top'].set_visible(False)

axarr[0, 0].spines['right'].set_visible(False)

axarr[0, 0].set_title('Axis [0,0]')



axarr[0, 1].scatter(X, Y)

axarr[0, 1].set_title('Axis [0,1]')

axarr[1, 0].plot(X, X ** 2)

axarr[1, 0].set_title('Axis [1,0]')

axarr[1, 1].scatter(X, X ** 1)

axarr[1, 1].set_title('Axis [1,1]')



plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определим квадратичную функцию
def quadratic_function(x):
    return 2 * x**2 - 5 * x + 3

# Зададим диапазон значений для x
x = np.linspace(-10, 10, 400)

# Вычислим значения функции
y = quadratic_function(x)

# Построим график
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$2x^2 - 6x + 3$', color='b')

# Настройки графика
plt.title('График квадратичной функции', fontsize=16)
plt.xlabel('Значения x', fontsize=12)
plt.ylabel('Значения f(x)', fontsize=12)
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```

```python
# Определим гармоническую функцию
def harmonic_function(x):
    return np.sin(x) + 0.5 * np.cos(2 * x)

# Зададим диапазон значений для x
x = np.linspace(0, 10 * np.pi, 500)

# Вычислим значения функции
y = harmonic_function(x)

# Построим график
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')

# Настройки графика
plt.title('График гармонической функции', fontsize=16)
plt.xlabel('Значения x', fontsize=12)
plt.ylabel('Значения f(x)', fontsize=12)
plt.grid(True)
plt.legend(loc='lower left')

# Показать график
plt.show()
```

```python
# Построим графики из п.2 и п.3 на разных подграфиках

# Создадим фигуру и зададим 2 подграфика
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Первый график: квадратичная функция
x1 = np.linspace(-10, 10, 400)
y1 = quadratic_function(x1)
ax1.plot(x1, y1, label=r'$2x^2 - 5x + 3$', color='b')
ax1.set_title('График квадратичной функции', fontsize=16)
ax1.set_xlabel('Значения x', fontsize=12)
ax1.set_ylabel('Значения f(x)', fontsize=12)
ax1.grid(True)
ax1.legend()

# Второй график: гармоническая функция
x2 = np.linspace(0, 10 * np.pi, 500)
y2 = harmonic_function(x2)
ax2.plot(x2, y2, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')
ax2.set_title('График гармонической функции', fontsize=16)
ax2.set_xlabel('Значения x', fontsize=12)
ax2.set_ylabel('Значения f(x)', fontsize=12)
ax2.grid(True)
ax2.legend(loc='lower left')

# Отобразим оба графика
plt.tight_layout()
plt.show()
```

```python
# Построим графики из п.2 и п.3 в одной координатной плоскости

# Первый график: квадратичная функция
x1 = np.linspace(-10, 10, 400)
y1 = quadratic_function(x1)

# Второй график: гармоническая функция
x2 = np.linspace(0, 10 * np.pi, 500)
y2 = harmonic_function(x2)

# Построим оба графика на одной координатной плоскости
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$2x^2 - 5x + 3$', color='b')
plt.plot(x2, y2, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')

# Настройки графика
plt.title('Графики квадратичной и гармонической функций', fontsize=16)
plt.xlabel('Значения x', fontsize=12)
plt.ylabel('Значения f(x)', fontsize=12)
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```

```python
# Определим параметрическую функцию
t = np.linspace(0, 4 * np.pi, 500)
x_parametric = np.cos(3 * t)
y_parametric = np.sin(2 * t)

# Построим график параметрической функции
plt.figure(figsize=(10, 6))
plt.plot(x_parametric, y_parametric, label=r'$x(t) = \cos(3t),\ y(t) = \sin(2t)$', color='purple')

# Настройки графика
plt.title('График параметрической функции', fontsize=16)
plt.xlabel('x(t)', fontsize=12)
plt.ylabel('y(t)', fontsize=12)
plt.grid(True)
plt.legend()

# Показать график
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные по прожиточному минимуму
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
min_income = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 
                       4900.79, 4792.54, 4806.32, 5358.53, 
                       6268.31, 7178.32])

# Построим круговой график
plt.figure(figsize=(8, 8))
plt.pie(min_income, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Настройки графика
plt.title('Доля прожиточного минимума по годам в Кыргызской Республике (2012-2022)', fontsize=16)
plt.axis('equal')  # Сделаем круговой график кругом

# Показать график
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем сетку координат
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

# Определяем произвольную функцию Z
Z = np.sin(np.sqrt(X**2 + Y**2))

# Построим каркасную поверхность
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='blue')

# Настройки графика
ax.set_title('Каркасная поверхность', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# Показать график
plt.show()
```
