```python
import numpy as np
import matplotlib.pyplot as plt

#   
def quadratic_function(x):
    return 2 * x**2 - 5 * x + 3

#     x
x = np.linspace(-10, 10, 400)

#   
y = quadratic_function(x)

#  
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$2x^2 - 6x + 3$', color='b')

#  
plt.title('  ', fontsize=16)
plt.xlabel(' x', fontsize=12)
plt.ylabel(' f(x)', fontsize=12)
plt.grid(True)
plt.legend()

#  
plt.show()
```

```python
#   
def harmonic_function(x):
    return np.sin(x) + 0.5 * np.cos(2 * x)

#     x
x = np.linspace(0, 10 * np.pi, 500)

#   
y = harmonic_function(x)

#  
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')

#  
plt.title('  ', fontsize=16)
plt.xlabel(' x', fontsize=12)
plt.ylabel(' f(x)', fontsize=12)
plt.grid(True)
plt.legend(loc='lower left')

#  
plt.show()
```

```python
#    .2  .3   

#     2 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

#  :  
x1 = np.linspace(-10, 10, 400)
y1 = quadratic_function(x1)
ax1.plot(x1, y1, label=r'$2x^2 - 5x + 3$', color='b')
ax1.set_title('  ', fontsize=16)
ax1.set_xlabel(' x', fontsize=12)
ax1.set_ylabel(' f(x)', fontsize=12)
ax1.grid(True)
ax1.legend()

#  :  
x2 = np.linspace(0, 10 * np.pi, 500)
y2 = harmonic_function(x2)
ax2.plot(x2, y2, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')
ax2.set_title('  ', fontsize=16)
ax2.set_xlabel(' x', fontsize=12)
ax2.set_ylabel(' f(x)', fontsize=12)
ax2.grid(True)
ax2.legend(loc='lower left')

#   
plt.tight_layout()
plt.show()
```

```python
#    .2  .3    

#  :  
x1 = np.linspace(-10, 10, 400)
y1 = quadratic_function(x1)

#  :  
x2 = np.linspace(0, 10 * np.pi, 500)
y2 = harmonic_function(x2)

#       
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$2x^2 - 5x + 3$', color='b')
plt.plot(x2, y2, label=r'$\sin(x) + 0.5\cos(2x)$', color='g')

#  
plt.title('    ', fontsize=16)
plt.xlabel(' x', fontsize=12)
plt.ylabel(' f(x)', fontsize=12)
plt.grid(True)
plt.legend()

#  
plt.show()
```

```python
#   
t = np.linspace(0, 4 * np.pi, 500)
x_parametric = np.cos(3 * t)
y_parametric = np.sin(2 * t)

#    
plt.figure(figsize=(10, 6))
plt.plot(x_parametric, y_parametric, label=r'$x(t) = \cos(3t),\ y(t) = \sin(2t)$', color='purple')

#  
plt.title('  ', fontsize=16)
plt.xlabel('x(t)', fontsize=12)
plt.ylabel('y(t)', fontsize=12)
plt.grid(True)
plt.legend()

#  
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

#    
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
min_income = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 
                       4900.79, 4792.54, 4806.32, 5358.53, 
                       6268.31, 7178.32])

#   
plt.figure(figsize=(8, 8))
plt.pie(min_income, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

#  
plt.title('        (2012-2022)', fontsize=16)
plt.axis('equal')  #    

#  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#   
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

#    Z
Z = np.sin(np.sqrt(X**2 + Y**2))

#   
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='blue')

#  
ax.set_title(' ', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

#  
plt.show()
```
